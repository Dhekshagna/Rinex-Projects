### Q19 . Plot a pie chart for displaying the England run percentage

```python
names=england_bat['Player']
runs=england_bat['runs']
plt.figure(figsize=(40,30))
plt.pie(runs,labels=names,autopct='%.2f%%')
plt.show()
```

### Q20. Plot a pie chart for displaying the India run percentage

```python
names=india_bat['Player']
runs=india_bat['runs']
plt.figure(figsize=(40,30))
plt.pie(runs,labels=names,autopct='%.2f%%')
plt.show()
```
